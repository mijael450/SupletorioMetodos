```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
trayNewton = []
trayBiseccion = []
def MetodoNewton(f, df, x0, tol):
    x = x0
    while abs(f(x)) > tol:
        x = x - f(x)/df(x)
        trayNewton.append(x)
    return x

def MetodoBiseccion(f, a, b, tol):
    while abs(f(a)) > tol:
        c = (a+b)/2
        if f(c) == 0:
            break
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
        trayBiseccion.append(c)
    return c


def f(x):
    return x**3 + 2*x**2-5

x = np.linspace(-3, 3, 600)
y = f(x)



#Raiz con metodo de Newton
root = MetodoNewton(f, lambda x: 3*x**2 + 4*x, 1, 1e-5)
print("Raiz x = ", root)


#Raiz con metodo de Bisecion    
root = MetodoBiseccion(f, -3, 3, 1e-5)
print("Raiz x = ", root)



#Grafica Comportamiento del metodo de Newton
print("Trayectoria Newton: ", trayNewton)
print("Trayectoria Biseccion: ", trayBiseccion)
print("Numero de iteraciones Newton: ", len(trayNewton))
print("Numero de iteraciones Biseccion: ", len(trayBiseccion))
#Criterio de parada utilizado en ambos metodos: tolerancia e iteracion maxima





plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.title("Metodo de Newton")
plt.plot(trayNewton, f(np.array(trayNewton)), 'ro-')
plt.grid()
plt.show()

```

```python
plt.title("Metodo de Biseccion")
plt.plot(trayBiseccion, f(np.array(trayBiseccion)), 'ro-')
plt.grid()
plt.show()
```
